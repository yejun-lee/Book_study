### 팬시 인덱싱 (29p)
리스트나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스에 해당하는 ndarray를 반환하는 인덱싱 방식.

```python
array1d = np.arange(start=1, stop=10)
array2d = array1d.reshape(3, 3)
print(array2d)

array3 = array2d[[0, 1], 2]     # 1행+2행 , 3열
print(array3.tolist())

array4 = array2d[[0, 1], 0:2]   # 1행+2행 , 1열+2열
# array4 = array2d[[0, 1], [0, 1]]
print(array4.tolist())

array5 = array2d[[0, 1]]        # 1행+2행
print(array5.tolist())
```

```python
array1d[[6, 7, 8]]  # 인덱스를 넣으면 인덱스에 해당하는 값 반환
```

### argsort() (35p)
정렬된 행렬의 인덱스를 반환하기

```python
org_array = np.array([3, 1, 9, 5])
sort_indices = np.argsort(org_array)
print(org_array)
print(sort_indices)   # 오름차순 정렬했을 때의 인덱스

sort_indices_1 = np.argsort(org_array)[::-1]
print(sort_indices_1) # 내림차순 정렬했을 때의 인덱스
```

```python
name_array = np.array(['A', 'B', 'C', 'D', 'E'])
score_array = np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)   # 오름차순 정렬했을 때 인덱스
print(sort_indices_asc)
print(name_array[sort_indices_asc])
print(score_array[sort_indices_asc])
```
